- 다형성  
상속 관계 내의 다른 클래스들의 인스턴스들이 같은 멤버 함수 호출에 대해 각각 다르게 반응하도록 하는 기능 // 같은 모양의 코드가 다르게 동작하는 것.

```python
class Game:
    def __init__(self):
        self.Strength = 10 # 힘
        self.Accuracy = 10 # 정확도
    
    def __str__(self):
        ability = f'''-----My ability-----
Strength (힘)  = {self.Strength}
Accuracy(정확) = {self.Accuracy}  '''
        return ability
    
    def levelUp(self):
        self.Strength += 1
        self.Accuracy += 1
        
    def __add__(self, n):
        self.Strength += n
        
    def __radd__(self, n):
        self.Strength += n + 5

class Warrior(Game): # 전사
    def levelUp(self):
        self.Strength += 5
        self.Accuracy += 2
    
    def __add__(self, n):
        self.Strength += n * 2
    
    def __radd__(self, n):
        self.Strength += n * 3
        
class Human(Game):   # 일반인
    pass

```

```python
# 기본스텟
h1 = Warrior()
h2 = Human()
print('[Warrior]\n', h1)
print()
print('[Human]\n', h2)
```

```python
# Game) 전사와 일반인은 같은 레벨업도 능력이 다르게 향상됨.
# code) 같은 levelUp() 함수를 호출했지만 다르게 적용됨
h1.levelUp()
h2.levelUp()

print('[Warrior]\n', h1)
print()
print('[Human]\n', h2)
```

3. 다음 각 요구사항 모두를 만족시키는 Counter 클래스를 코딩하시오

```python
class Counter:
    #요구사항 1
    def __init__(self, count, step = 1):
        self.count = count
        self.step = step
    
    #요구사항 2
    def __str__(self):
        return f"[Count (step: {self.step})] {self.count}"
    
    #요구사항 3
    def incr(self):
        self.count += self.step
    
    #요구사항 4
    def __call__(self):
        self.incr()
    
    #요구사항 5
    def __add__(self, other):
        num = self.count + other
        return Counter(num, self.step)
    
    def __sub__(self, other):
        num = self.count - other
        return Counter(num, self.step)
    
    #요구사항 6
    def __eq__(self, other):
        return self.count == other
    
    def __lt__(self, other):
        return self.count < other
    
    def __gt__(self, other):
        return self.count > other
    
    def __le__(self, other):
        return self.count <= other
    
    def __ge__(self, other):
        return self.count >= other
```

4. 다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __ init __(), __ str() __(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

```python
class MySet(list):
    def __init__(self, l): # 객체가 생성될때 자동으로 호출되는 함수
        for e in l: # 받은 list [1,2,2,3]
            self.append(e) # 부모 클래스 list의 append메서드를 이용해서 e를 append
        MySet.eliminate_duplicate(self) # MySet클래스의 elim함수 호출
    
    def __str__(self): # print 하면 __str__이 반응
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ," # self의 원소를 result 뒤에 하나씩 붙여준다
        result = result[0:len(result)-2] + "}" # 끝에  ,를 제거하고 마무리로 } 붙임  
        return result # result 리턴

    @staticmethod # 정적 메소드   
    def eliminate_duplicate(l): # 인스턴트 객체 없이 클래스에서 직접 호출 가능
        s = []
        for e in l:
            if e not in s: # s안에 e가 없으면 즉, 중복은 넣지 않는다
                s.append(e) # s에도 append 해준다 s = [1,2,3]
        l[:] = [] # l 리스트 초기화
        for e in s: 
            l.append(e) # l = [1,2,3]

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

#1 __ init __  
- 생성자라고 불리며, 인스턴트 객체가 self, 인자가 l가 됩니다.  
- 생성자는 객체가 생성될때 자동으로 호출되는 함수로 객체가 생성되면 아래 과정 대로 작동합니다.  
  1) 인스턴스 객체 생성하면서 인자로 [1,2,2,3] 을 받습니다.  
  2) 객체가 생성되면서 __ init __ () 함수가 호출됩니다.  
  3) 인자로 받은 시퀀스 자료형의 원소를 for ~ in을 활용하여 하나씩 뽑아 내고,  
  3) list를 서브클래싱 했기 때문에 list클래스의 append메서드를 활용해 self에 원소 하나씩 append 해줍니다.   
  4) MySet 클래스 내의 eliminate_duplicate() 함수를 직접 호출합니다. 이때 인자를 자기자신인 self를 넘겨 줍니다. ([1,2,2,3])  

#2 __ str __
- print(class)를 하게 되었을 때 반응하는 매직 메서드 입니다.  
- result를 반환하고 result는 "MySet: { 받은 인자의 원소들 }" 입니다.
- 예를들어 self = [1,2,2,3]으로 인자를 받아왔고, for 문을 다 돌고나면,  
result = "MySet : {1, 2, 2, 3}" 이 되고 이를 반환합니다.  

#3 eliminate_duplicate()  
- 이 클래스의 핵심 함수로, 내장 자료형 set 처럼 중복을 제거해주는 함수 입니다.  
- 이 함수는 위에 장식자로 @staticmethod 되어 있고 정적 메소드가 됩니다.  
- 정적 메소드는 이름 공간과 상관없이(인스턴트 객체 없이) 클래스에서 직접 호출 가능합니다.  
- 그래서 __ init __ 에 바로 MySet이라는 클래스의 함수 호출이라는 형식으로 사용합니다.
- 결과부터 이야기 하면 예를들어 [1,2,2,3] 이 들어가면 중복이 사라진 [1,2,3]가 나옵니다.  
- 그 과정은 다음과 같습니다.  
  1) 인자로 self를 넘겨주고 l로 받으면서 함수가 호출됩니다.  
  2) s 라는 객체에 빈 리스트([]) 를 할당합니다.  
  3) 인자로 받은 객체의 원소를 하나씩 e에 할당하면서, 만약 그 원소 e가 새로 만든 s라는 리스트에 없다면 s에 그 원소를 append 합니다.  
  
  예를들어 [1,2,2,3]이 들어온다면 s는 [] -> [1] -> [1,2] -> [1,2,3] 이 됩니다.  
  
  4) 인자로 받은 리스트를 빈리스트로 초기화 합니다.  
  5) 이제 반대로 s의 원소를 모두 인자로 받은 l 리스트에 append 합니다.

5. MySet 클래스에 메소드를 추가(입력 x -> 메소드만 만들어서 주석)

```python
class MySet(list):
    def __init__(self, l): # 객체가 생성될때 자동으로 호출되는 함수
        for e in l: # 받은 list [1,2,2,3]
            self.append(e) # 부모 클래스 list의 append메서드를 이용해서 e를 append
        MySet.eliminate_duplicate(self) # MySet클래스의 elim함수 호출
    
    def __str__(self): # print 하면 __str__이 반응
        result = "MySet: {"
        for e in self:
            result = result + str(e) + ", " # self의 원소를 result 뒤에 하나씩 붙여준다
        result = result[0:len(result)-2] + "}" # 끝에  ,를 제거하고 마무리로 } 붙임  
        return result # result 리턴

    @staticmethod # 정적 메소드   
    def eliminate_duplicate(l): # 인스턴트 객체 없이 클래스에서 직접 호출 가능
        s = []
        for e in l:
            if e not in s: # s안에 e가 없으면 즉, 중복은 넣지 않는다
                s.append(e) # s에도 append 해준다 s = [1,2,3]
        l[:] = [] # l 리스트 초기화
        for e in s: 
            l.append(e) # l = [1,2,3]
    
    def __or__(self, other):        # 클래스의 인스턴스가 | 좌측에 있을 때 호출
        return MySet(self + other)  # 두 리스트를 합쳐 -> MySet 클래스에 인자로 넘겨줘
                                    # 중복이 제거 된 MySet을 반환
    
    def __and__(self, other): # 클래스의 인스턴스가 & 좌측에 있을 때 호출
        s = []                # 빈리스트 만들고
        for e in self:        # self의 원소하나씩 e에 할당해
            if e in other:    # 만약 other에 있으면
                s.append(e)   # s에 append
        return MySet(s)       # s를 MySet 클래스에 인자로 주고 반환
    
    def __sub__(self, other): # 클래스의 인스턴스가 - 좌측에 있을 때 호출
        for e in other:       # 인자로 받은 other의 원소 하나씩 e에 할당하고
            if e in self:     # 만약 원소 e가 self에 있다면
                self.remove(e)# self에서 그 원소를 remove 
        return MySet(self)    # self를 MySet 클래스 인자로 주고 반환
        
```

```python
class SearchEngine:
    def __init__(self, *args):
        self.url = [i for i in args]
        self.dict = [self.make_dict(url) for url in self.url]
        self.top = [self.make_top(d) for d in self.dict]
    
    def update(self, url):
        dict1 = self.make_dict(url)
        self.dict = [self.make_dict(url) for url in self.url]
        self.top = [self.make_top(d) for d in self.dict]
    
    def addUrl(self, url):
        self.url.append(url)
        self.update(url) # 갱신
    
    def removeUrl(self, url):
        self.url.remove(url)
        self.update(url) # 갱신
        
    def listUrls(self):
        for url in self.url:
            print(url)
    
    def getWordsFrequency(self):
        if self.url == []: return {}
        else:              return self.dict
        
    def getMaxFreqencyWords(self):
        if self.dict == []: return None
        else:               return self.top
    
    def searchUrlByWord(self, word):
        self.rank(word)
    
    
    
    #self.dict 만들기
    def make_dict(self, url):
        #url source 가져오기
        req = requests.get(url) #게임
        source = req.text
    
        # source 관리 
        source = source.replace('\n','') # 들여쓰기 제거
        source = source.replace('\t','') # Tap 제거
        source = ' '.join(source.split()) # 중복 공백문자 제거

        front_index, back_index, str_start, str_end = 0, 0, 0, 0

        index = 0
        str_list = []

        switch = 1 # script 제거하기 위한 스위치 / 1 - On , 0 - Off

        for i in source:

            #인지 부분
            if i == '<':
                front_index = index + 1      # < > 안의 문자확인을 위한 인덱스 저장

                # script, style, url 를 만나면 스위치=0 OFF
                if source[front_index : front_index + 6] == 'script': switch = 0
                elif source[front_index : front_index + 5] == 'style': switch = 0
                elif source[front_index : front_index + 3] == 'url': switch = 0


            elif i == '>':
                back_index = index       # < > 안의 문자확인을 위한 인덱스 저장

                # /script, /style, /url를 만나면 스위치=1 ON
                if source[back_index - 8 : back_index + 1] == '</script>': switch = 1
                elif source[back_index - 7 : back_index + 1] == '</style>': switch = 1
                elif source[back_index - 5 : back_index + 1] == '</url>': switch = 1

                elif switch == 1 : str_start = back_index + 1
                # 스위치 ON 이고 '>'만났기에 문자열 시작으로 그 때의 인덱스 저장    




            if str_start != 0 and i == '<':  # 스위치 0n 이고 '<' 만나면
                str_end = index # 문자열 끝이므로 그 때의 인덱스 저장

                #삽입 부분
                if source[str_start : str_end] != '' and source[str_start : str_end] != ' ':  # 비어있는 값 삭제
                    str_list.append(source[str_start : str_end]) # 시작 인덱스 ~ 끝 인덱스

                #초기화
                str_start = 0
                str_end = 0        

            index += 1


        # 추가 부분 : &lt, &gt, &middot 도 없애기
        for s in str_list:
            if s[0] == '&':
                str_list.remove(s)
        
        for s in str_list:
            if '&nbsp' in s:
                str_list.remove(s)





        # 구두문자 제거 부분
        for i in range(len(str_list)):
            str_list[i] = str_list[i].translate(str.maketrans('','', string.punctuation))

        true_str =[]
        f_index = 0

        #텍스트 문자열 -> 순수 텍스트 문자열
        for str1 in str_list:
            switch = 1
            for i in range(len(str1)):
                if switch != 0 and str1[i] != ' ':
                    f_index = i
                    switch = 0

                elif str1[i] == ' ':
                    true_str.append(str1[f_index:i])
                    switch = 1

                if i == len(str1) - 1:
                    true_str.append(str1[f_index:i+1])

        # 순수 텍스트 문자열 -> 순수 텍스트 빈도 dict
        str_dict = dict()

        for i in range(len(true_str)):
            if true_str[i] not in str_dict: str_dict[true_str[i]] = 1
            else:
                str_dict[true_str[i]] = str_dict.get(true_str[i]) + 1

        # 구두문자 제거하면서 생긴 빈 '' 값 제거
        if '' in str_dict.keys():
            del(str_dict[''])

        # 불용어 제거
        self.bool_del(str_dict)

        return str_dict


    # 불용어 제거 함수
    def bool_del(self, str_dict):
        with open('stop_words_english.txt', 'rt', encoding='UTF8') as f:
            bool_list = f.read().split()

        with open('stop_words_korean.txt', 'rt', encoding='UTF8') as f:
            bool_list2 = f.read().split()

        # 영어 제거
        for str1 in bool_list:
            if str1 in str_dict.keys():
                #print(str1)
                del(str_dict[str1]) 

        # 한글 제거
        for str2 in bool_list2:
            if str2 in str_dict.keys():
                #print(str2)
                del(str_dict[str2])
    
    #self.top 만들기
    def make_top(self, dict1):
        top_list, word_list = [], []
        for key, value in dict1.items():
            word_list.append((key, value))
        word_list.sort(key = lambda x: x[1], reverse = True)

        top_list.append(word_list[0])
        for i in range(1, len(word_list)):
            if top_list[0][1] == word_list[i][1]:
                top_list.append(word_list[i])
            else: break
        
        return top_list
                    
    def rank(self, word):
        rank_score = [0.1] * len(self.dict)
        
        for i in range(len(self.dict)):
            if word in self.dict[i].keys():
                rank_score[i] += self.dict[i][word]
        
        n = max(rank_score)
        index = rank_score.index(n)
        while True:
            print(self.url[index])
            rank_score[index] = 0
            
            if max(rank_score) == n:
                n = max(rank_score)
                index = rank_score.index(n)
            else: break
```

2. 위 1번 문제에서 정의한 SearchEngine 클래스를 상속하여 SearchEngineWithOrderedWebWords 클래스를 정의하고 슈퍼클래스에 정의된 getWordsFrequency() 메소드를 오버라이드하여 단어 출현 빈도를 내림 차순으로 정렬하여 리스트로 출력하시오.

```python
class SearchEngineWithOrderedWebWords(SearchEngine):
    def getWordsFrequency(self, reverse = False):
        dict1 = SearchEngine.getWordsFrequency(self)
        top_list, self.word_list = [], []
        
        for d in dict1:
            a = [(key, value) for key, value in d.items()]
            a.sort(key = lambda x: x[1], reverse = reverse)
            self.word_list.append(a)
        
        return self.word_list
    
    def __iter__(self):
        self.count = -1
        return self
    
    def __next__(self):
        n = [len(w_list) for w_list in self.word_list]
        if min(n) - 1 == self.count:
            raise StopIteration
        else:
            self.count += 1
            if len(self.word_list) == 1:
                return self.word_list[0][self.count]

            else:
                return [w_list[self.count] for w_list in self.word_list]       
```

```python
w4 = SearchEngineWithOrderedWebWords('http://www.times.com','https://github.com')
```
