5. 다음과 같이 0보다 큰 정수 리스트 변수 list를 인자로 받는 함수 addall(list)와 addallodd(list)를 for ~ in 리터널과 리스트 내포 방식으로 각각 제시하시오.  
- 1) addall(list) 함수 (리스트 내의 모든 정수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)  
  
  
- 2) addallodd(list) 함수 (리스트내의 모든 홀수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)


- [참고 1]: 리스트 내포 방식으로 addall() 및 addallodd()함수를 만들 때엔 리스트 내포에 의해 생성된 리스트 결과에 대해 내장 함수를 사용하는 방식을 고려해보시오 (주의: 리스트 내포 방식의 addall() 및 addallodd() 함수의 몸체는 단 1줄로 작성되어야 한다.)
- [참고 2]: sum() 등의 내장함수 사용 가능함

```python
a = 0
```

```python
#1) for ~ in 리터널

def addall(alist):    
    a = 0
    for i in alist:
        a = a + i
    return a

addall([1,2,3,4,5,6,7,8,9])
```

```python
#1) 리스트 내포
def addall(alist):
    return sum([i for i in alist])

addall([1,2,3,4,5,6,7,8,9])
```

```python
#2) for ~ in 리터널

def addallodd(alist):
    a = 0
    
    for i in alist:
        if i % 2 == 1: a += i
    
    return a

addallodd([1,2,3,4,5,6,7,8,9])
```

```python
#2) 리스트 내포

def addallodd(alist):
    return sum([i for i in alist if i % 2 == 1])

addallodd([1,2,3,4,5,6,7,8,9])
```

8. 다음 문제를 정확히 읽고 적합한 코딩을 제시하시오.
 - length(X)를 다음과 같이 정의한다 : 10진수로 표현된 X의 자릿수
 - Strange라는 숫자를 아래과 같이 정의한다.
    - 음이 아닌 정수 중 자릿수가 1인 모든 숫자는 Strange이다.
    - length(X) > 1 인 숫자 X 는 다음 2가지를 만족하면 Strange이다.
        - X는 length(X)로 정확히 나누어 떨어진다.
        - X / length(X) 가 strange이다.  
          
          
 - 이 문제에서 숫자 하나가 주어질 때, 이 숫자가 Strange인지 판단하는 함수 isStrange(num)을 작성하시오.
 - isStrange(num) 함수가 재귀적인 함수가 되도록 작성해 보세요.

```python
def length(X): 
    if X < 0: return len(str(X)) -1
    return len(str(X))


def isStrange(num):
    
    if length(num) == 1 and num < 0: return False
    
    elif num % length(num) != 0: return False
    
    elif length(num) != 1: isStrange(num / length(num))
        
    return True
    

num = 56
print(f"num = {num}                   =>   {isStrange(num)}")
```

```python
print(f"num = {0}                     =>   {isStrange(0)}")
print(f"num = {56}                    =>   {isStrange(56)}")
print(f"num = {100}                   =>   {isStrange(100)}")
print(f"num = {7882602506492313600}   =>   {isStrange(7882602506492313600)}")
print(f"num = {6012302549522098651}   =>   {isStrange(6012302549522098651)}")
```
